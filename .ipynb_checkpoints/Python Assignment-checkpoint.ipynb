##   Q-1 String Operations

##  Q-2 Numeric Data Types and Conversion Function

```python
# Get the input number from the user
num = input("Enter a number: ")

# Convert the input to different numeric data types
int_num = int(num)  # Convert to integer
float_num = float(num)  # Convert to float
complex_num = complex(num)  # Convert to complex

# Display the converted values
print("Integer:", int_num)
print("Float:", float_num)
print("Complex:", complex_num)

# Explanation of the difference between these data types:
# 
# **Integer (int)**: A whole number, either positive, negative, or zero, without a fractional part. 
#                    Examples: 1, -2, 0
# 
# **Float (float)**: A number with a fractional part, which can be positive, negative, or zero. 
#                    Examples: 3.14, -0.5, 0.0
# 
# **Complex (complex)**: A number that consists of a real part and an imaginary part, 
#                        which can be expressed in the form a + bj, where a and b are real numbers 
#                        and j is the imaginary unit, which satisfies j^2 = -1. 
#                        Examples: 3 + 4j, -2 - 3j, 0 + 0j
```

## Q-4 Using the Format Method

```python
# Prompt the user to enter the length and width of the rectangle
length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))

# Calculate the area
area = length * width

# Format the output to display the area with two decimal places
output = "The area of the rectangle is: {:.2f}".format(area)

# Display the result
print(output)
```

## Q-5 The % Method and print Function

```python
# Prompt the user to enter three numbers
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))

# Calculate the average
average = (num1 + num2 + num3) / 3

# Use the % method for string formatting to display the average
output = "The average of the three numbers is: %.2f" % average

# Display the result
print(output)
```

## Q-6 Control Flow (if Statements and Loops)

```python
# Create an infinite loop that continues until the user enters 'exit'
while True:
    # Prompt the user to enter a number or 'exit' to quit
    user_input = input("Enter a number (or 'exit' to quit): ")

    # Check if the user wants to exit
    if user_input.lower() == 'exit':
        # Use the break statement to exit the loop
        break

    # Try to convert the input to a float
    try:
        num = float(user_input)
    except ValueError:
        # If the input is not a valid number, prompt the user to try again
        print("Invalid input. Please enter a number.")
        continue

    # Determine whether the number is positive, negative, or zero
    if num > 0:
        print("The number is positive.")
    elif num < 0:
        print("The number is negative.")
    else:
        print("The number is zero.")

    # Continue to the next iteration of the loop
    continue
```

## Q-7 Relational and Logical Operators

```python
# Prompt the user to enter two numbers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Use relational operators to determine if each number is even or odd
is_num1_even = num1 % 2 == 0
is_num2_even = num2 % 2 == 0

# Use logical operators to determine the relationship between the two numbers
if is_num1_even and is_num2_even:
    print("Both numbers are even.")
elif not is_num1_even and not is_num2_even:
    print("Both numbers are odd.")
else:
    print("One number is even and the other is odd.")
```

## Q-8 For Loop and Bitwise Operators

```python
# Prompt the user to enter an integer
num = int(input("Enter an integer: "))

# Define a list of bases to convert to (binary, octal, hexadecimal)
bases = ["binary", "octal", "hexadecimal"]

# Define a list of corresponding format strings
formats = ["b", "o", "x"]

# Use a for loop to iterate over the bases and formats
for base, fmt in zip(bases, formats):
    # Use the format() function to convert the integer to the desired base
    # The "#" flag is used to include the base prefix (e.g. "0b" for binary)
    result = format(num, "#" + fmt)

    # Print the result
    print(f"The {base} equivalent of {num} is {result}")
```
